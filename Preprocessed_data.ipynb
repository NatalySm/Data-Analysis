#### 2.2 Добавление новых колонок (дубликат)

```python
df['floor_q'] = df['floor']
df['price_m2'] = df['price']
df['street number'] = df['street']
df['elevator'] = df['floor']
```

##### 2.3. Поставить названия колонок в нужном порядке

```python
df = df[['city_municipality', 'city_parish', 'floor', 'floor_q', 'area','comfortably','free_text','rooms_q','price','price_m2','type','serial','street','street number','elevator','trade_check']]
```

#### 2.4. Очистка данных от ненужных значений

```python
#Убераем все что после первого символа
df['floor'] = df['floor'].str[:1]
#Убераем все что до второго символа
df['floor_q'] = df['floor_q'].str[2:]
#Удаляем /лифт и /
df['floor_q'] = df['floor_q'].str.replace("/лифт", '')
df['floor_q'] = df['floor_q'].str.replace("/", '')
```

```python
#удаление все что в скобках ()
df['price'] = df['price'].str.replace(r"\(.*?\)", "")
```

```python
#удаление символа €
df['price'] = df['price'].str.replace(" €", '')
df['price'] = df['price'].str.replace(" ", '')
df['price_m2'] = df['price_m2'].str.replace(" ", '')
df['price_m2'] = df['price_m2'].str.replace(" €", '')
```

```python
#удаление все что до скобок
df['price_m2'] = df['price_m2'].str.split('(').str[1]
```

```python
#удаление лишних символов
df['price_m2'] = df['price_m2'].str.replace("€/м²", '')
df['price_m2'] = df['price_m2'].str.replace((')'), '')
```

```python
# замена минуса на -1 этаж
df['floor'] = df['floor'].str.replace("-", '-1')
```

```python
# замена минуса на 0(так как не известно сколько комнат)
df['rooms_q'] = df['rooms_q'].str.replace("-", '0')
```

```python
# если значение равно "лифт", то ставиться True, иначе False
df['elevator'] = np.where(df['elevator'].str.contains('лифт'), True, False)
```

```python
# переводим колонку "elevator"  в тип int
df['elevator'] = df['elevator'].astype(int)
```

```python
# переводим колонку "price" и "price_m2"  в тип float
df['price'] = df['price'].astype(float)
df['price_m2'] = df['price_m2'].astype(float)
```

```python
df.dtypes
```

```python
#если в колонке 'floor_q' значение этажей больше, чем 5 , то ставим в колонке 'elevator' значение 1
df['elevator'] = np.where(df['floor_q'] > 5, 1, 0)
```

```python
#
df['street'] = df['street'].str.split()
df['street number'] = df['street number'].str.split()
```

```python
def extract_house_number(row):
    return row['street'][-1]
```

```python
df['house_number'] = df.apply(extract_house_number, axis=1)
```
